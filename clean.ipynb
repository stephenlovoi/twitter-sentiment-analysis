```python
import pandas as pd
path = 'Resources/trump_archive.csv'
df = pd.read_csv(path)
```

```python
# Drop unnecessary columns, drop retweeted columns
df = df.drop(['id', 'isDeleted', 'device', 'isFlagged'], axis=1)
df = df.loc[df['isRetweet'] != 't']
```

```python
# Drop rows with links 
df = df.loc[df['text'].str.contains('https') == False]
```

```python
df.to_csv('Resources/trump_archive.csv')
```
