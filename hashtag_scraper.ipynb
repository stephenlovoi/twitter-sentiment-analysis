```python
# insert hashtag or keyword in 'search_words' to get results you are looking for
search_words = "#TrumpWall"
date_since = "2021-01-01"

username = []
location = []
text = []
hashtags = []
time = []

for i in tw.Cursor(api.search, q=search_words+' -filter:retweets', lang="en", since=date_since, tweet_mode='extended').items(3200):
    username.append(i.user.screen_name)
    location.append(i.user.location)
    text.append(i.full_text)
    hashtags.append(i.entities['hashtags'])
    time.append(i.created_at)
```
